# Questions


<h3>Which algorithm or algorithms would be most appropriate for planning in a very restricted domain (i.e., one that has only a few actions) and needs to operate in real time?</h3>

<h4>Answer:</h4>

The greedy best first searches algorithms, regardless of heuristics, reliably performed the fast enough for real time operations when action states were kept low.

<h3>Which algorithm or algorithms would be most appropriate for planning in very large domains (e.g., planning delivery routes for all UPS drivers in the U.S. on a given day)</h3>

<h4>Answer:</h4>

Breadth First Search was the more optimal at exploring nodes and expanding the graph. A\* searches worked well also at exploring large domains at the trade off for computational expense.

<h3>Which algorithm or algorithms would be most appropriate for planning problems where it is important to find only optimal plans?</h3>

<h4>Answer:</h4>

All A\* Search variants and BFS reliably found the most optmial plan.
